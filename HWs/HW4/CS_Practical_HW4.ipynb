```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
# This function generates poisson numbers based on acceptance-rejection method

def poisson_generator(_lambda,n):
    if _lambda<15:
        for _ in range(n):
            p=1
            i=0
            boundary=np.e**(-_lambda)
            while True:
                p*=np.random.rand()
                if p<boundary:
                    break
                i+=1
            yield i
    else:
        for _ in range(n):
            s=0
            i=0
            while True:
                s+=np.log(np.random.rand())
                if s<-_lambda:
                    break
                i+=1
            yield i 
```

```python
# Count of numbers generated by poisson_generator
count_numbers=100000

# Poisson distribution lambda parameter
_lambda=6
numbers=list(poisson_generator(_lambda=_lambda,n=int(count_numbers)))
```

```python
plt.hist(numbers)
plt.title('Generated numbers histogram')
plt.show()
```

```python
# Plot the poisson distribution
def plot_poisson(ax,_lambda,max_num,**args):
    pdf_x = np.arange(0,max_num,dtype=np.int32)
    pdf_y=poisson.pmf(k=pdf_x,mu=_lambda)
    ax.plot(pdf_x,pdf_y,**args)
    ax.set_title('Poisson PDF')
```

```python
fig,ax=plt.subplots()
plot_poisson(ax,_lambda,max_num=np.max(numbers))
```

```python
fig,ax=plt.subplots()
plot_poisson(ax,_lambda,max_num=np.max(numbers),label='Real poisson distribution')
ax.hist(numbers,density=True,label='Generated numbers distribution')
ax.legend()
ax.set_title('Generated poisson numbers\' histogram vs poisson\'s PDF')
ax.plot()
pass
```
